```python
# put it all together
def lengthOfLongestSubstring(s):
    substring_list = []
    for counter, char in enumerate(s):
        new_s = ''
        for i in range(counter, len(s)):
            if s[i] not in new_s:
                new_s += s[i]
            else:
                break
        substring_list.append(new_s)
    length_list = [len(substring) for substring in substring_list]
    max_length = max(length_list)
    return max_length
```

## Roman to Integer (Easy) 

```python
s = '123'
```

**Approach**: Look at each digit in the numeral and consider all possible cases uses if, elif, else logical statements. 

Possible cases to consider in logic:
- M (only changes if preceded)
    - First numeral = 1000
    - Anywhere else and preceded by C = 900
    - Anything else = 1000
- D (only changes if preceded)
    - First numeral = 500
    - Anywhere else and preceded by C = 400
    - Anything else = 500
- C (changes if preceded or followed)
    - Last numeral = 100 (not needed?)
        - Unless preceded by X, = 90
    - First numeral = 100
    - Before M or D = 0
    - Preceded by X = 90
    - Else = 100
- L (changes if preceded)
    - First numeral = 50
    - Preceded by X = 40
    - Else = 50
- X (changes if preceded or followed)
    - Last numeral = 10 (not needed?)
        - Unless preceded by X, = 9
    - First numeral = 10
    - Before L or C = 0
    - Preceded by I = 9
    - Else = 10
- V (changes if preceded)
    - First numeral = 5
    - Preceded by I = 4
    - Else = 5
- I (changes if followed)
    - Last numeral = 1 (not needed?)
    - Before V or X = 0
    - Else = 1

```python
# first version, doesn't work for all cases, i.e. "IX"
def romanToInt(s):
    number = 0
    for position, char in enumerate(s):
        if char == 'M':
            if position == 0:
                number += 1000
            elif s[position-1] == 'C':
                number += 900
            else:
                number += 1000
        if char == 'D':
            if position == 0:
                number += 500
            elif s[position-1] == 'C':
                number += 400
            else:
                number += 500
        if char == 'C':
            if position == 0:
                if s[position+1] == 'M' or s[position+1] == 'D':
                    number += 0
                else:
                    number += 100
            elif position == (len(s)-1):
                if s[position-1] == 'X':
                    number += 90
                else:
                    number += 100
            elif s[position+1] == 'M' or s[position+1] == 'D':
                number += 0
            elif s[position-1] == 'X':
                number += 90
            else:
                number += 100
        if char == 'L':
            if position == 0:
                number += 50
            elif s[position-1] == 'X':
                number += 40
            else:
                number += 50
        if char == 'X':
            if position == 0:
                if s[position+1] == 'L' or s[position+1] == 'C':
                    number += 0
                else:
                    number += 10
            elif position == (len(s)-1):
                if s[position-1] == 'I':
                    number += 9
                else:
                    number += 10
            elif s[position+1] == 'L' or s[position+1] == 'C':
                number += 0
            elif s[position-1] == 'I':
                number += 9
            else:
                number += 10
        if char == 'V':
            if position == 0:
                number += 5
            elif s[position-1] == 'I':
                number += 4
            else:
                number += 5
        if char == 'I':
            if position == (len(s)-1):
                number += 1
            elif s[position+1] == 'V' or s[position+1] == 'X':
                number += 0
            else:
                number += 1
    return number
```

**Approach**: Look for special two numeral sequences, calculate their quantity, remove them, and the sum up quantities for remaining numerals and add them

```python
# another approach that is much more clean and works for all cases
def romanToInt(s):
    # create a dictionary with key being numerals and value being quantity associated with each
    # include the special two-character combinations
    value_roman = {"M":1000, "CM":900, "D":500, "CD": 400,
                   "C":100,"XC":90, "L":50, "XL":40,
                   "X":10, "IX":9, "V":5, "IV":4, "I":1}
    # initiate variable to hold final integer translation of roman numerals
    num = 0
    # create a list of the just the special two-numeral combinations
    specials_list = ["CM","CD","XC","XL","IX","IV"]
    # loop through special list
    for i in specials_list:
        # Check if specials are contained in roman numeral being translated
        if i in s:
            # if special is present, add to integer translation using dictionary to translate
            num += value_roman[i]
            # also remove the special from the roman numeral being trasnlated to leave behind remaining numerals
            s=s.replace(i,"")
    # loop through remaining numerals
    for i in s:
        # Add integer translation using dictionary to translate
        num += value_roman[i]
    # return final number
    return(num)
```

Works for all.

## Longest Common Prefix (Easy)

```python
def longestCommonPrefix(strs):
    shortest = min([len(str) for str in strs]) # find the length of the shortest string and only iterate through that number of characters for each string
    temp = '' # create variable to hold combination of ith characters for each string in list for each iteration 
    prefix = '' # create variable to hold only those characters that are in common for all strings in list (common prefix)
    for char in range(shortest): # create for loop to iterate through each character up to length of shortest string (longest possible common prefix)
        for word in strs: # create for loop to iterate through each string in list
            temp += word[char] # add char for each word
        if temp == len(temp) * temp[0]: # check if all chars are the same for all words
            prefix += word[char] # if the same, add to prefix
        else: # add else condition so that we are only looking at prefixes and not all common letters in words
            break # break the initial for loop and go straight to return statement if all letters don't match since this signals end of common prefix
        temp = '' # reset temp variable for each character position
    return prefix
```
