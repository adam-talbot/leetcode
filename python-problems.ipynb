## Longest Substring Without Repeating Characters (Medium)

```python
s = 'abcabcbb'
```

```python
s
```

```python
for char in s:
    print(char)
```

```python
if 'a' in s:
    print(True)
```

```python
len(s)
```

```python
new_s = ''
for char in s:
    if char not in new_s:
        new_s += char
print(new_s)
```

```python
list(s)
```

```python
# iterate through string
# add each character to new string
# check when adding each new character if the string already contains the new character
# if if does, stop adding characters
# if it doesn't, keep adding
# save length of substring
# start at every index of string and return longest substring of all substrings saved in list of lists
```

```python
# now do this starting at every index in the list, saving each substring to a list of lists, and the pulling out the longest and returning it
print(s)
substring_list = []
for counter, char in enumerate(s):
    print(counter, char)
    new_s = ''
    for i in range(counter, len(s)):
        print(s[i])
        if s[i] not in new_s:
            new_s += s[i]
    substring_list.append(new_s)
    print(new_s)
    print(substring_list)
    print('---')
```

```python
max(substring_list)
```

```python
substring_list
```

```python
length_list = [len(substring) for substring in substring_list]
```

```python
length_list
```

```python
max(length_list)
```

```python
# put it all together
def lengthOfLongestSubstring(s):
    substring_list = []
    for counter, char in enumerate(s):
        new_s = ''
        for i in range(counter, len(s)):
            if s[i] not in new_s:
                new_s += s[i]
            else:
                break
        substring_list.append(new_s)
    length_list = [len(substring) for substring in substring_list]
    max_length = max(length_list)
    return max_length
```

## Palindrome Number (Easy)

```python
x = 123
```

```python
x_str = str(x)
```

```python
x_str
```

```python
x_str[::-1]
```

```python
def isPalindrome(x):
    x_str = str(x)
    if x_str == x_str[::-1]:
        return True
    else:
        return False
```

```python
isPalindrome('123')
```

```python
isPalindrome('121')
```
